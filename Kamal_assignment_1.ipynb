```python
  mylist = [3, 4, 5, 3, 2, 6, 6, 7, 6, 7]

  #create a dictionary to add its key & value ... eg 3:1, 6:2
  occur = {}

  #loop items in the list one at a time which are called "i"
  for i in mylist:

  #if i exists in dict occur{}, increment by 1, becomes 3:2, 6:2 etc
    if i in occur:
      occur[i] = occur[i] + 1

  #else  when it is not a key in occur, add it with value=1, so becomes 3:1, 4:1 etc
    else:
      print("item = ", i)
      occur[i] = 1
  print("Frequency of each item = ",occur)
  print(occur.items())
```

```python
# another example
mylist = [3,3,2,4,2,2]
occur = {}

for i in mylist:
  if i in occur:
    occur[i] = occur[i] + 1
    print("  Added in dict",occur)
  else:
    occur[i] = 1
    print("First time",occur)

print("Occur dict with frequency",occur)
```

## **Create a dictionary that stores student names as keys and marks as values, then find the student with the highest marks.**

```python
students = {'Raja':85, 'John':25, 'Siva':90}
#students.values()
print(students['Siva'])

# assume that 25 is the largest
largest = 0
s_name = None

for name,marks in students.items():
  if marks > largest:
    largest = marks
    s_name = name
print(s_name,largest)



```
