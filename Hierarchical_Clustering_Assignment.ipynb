```python
df = data[['Settlement size','Income']]
```

```python
x = df.values
```

```python
x
```

```python
# Dendrogram
import scipy.cluster.hierarchy as den
import matplotlib.pyplot as plt

plt.figure(figsize=(20,7))
plt.title("Dendogram")

den.dendrogram(den.linkage(x,method="ward"))
plt.ylabel("Eucledian distance")
plt.xlabel("Attr")
plt.show()
```

##Training the Dataset and predicting the best Optimal no. of clusers = 4

```python
from sklearn.cluster import AgglomerativeClustering

Hierarchical_Model = AgglomerativeClustering(n_clusters=4,linkage="ward")
y = Hierarchical_Model.fit_predict(x)
```

```python
y
```

##Calculating the accuracy of the trained model

```python
from sklearn.metrics import silhouette_score
```

```python
score = silhouette_score(x,y)
```
