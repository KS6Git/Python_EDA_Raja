```python
# Get position of an item in the list
list_4 = [4,5,6,34,546,90,0.5,'monk']
print(list_4[7])

c = 0
for i in list_4:
  if i == 'monk':
    print(c)
  c = c + 1
```

```python
# Look for all positions of 6
list_4 = [4,5,6,34,546,90,0.5,6,'monk']

c = 0
for i in list_4:
  if i == 6:
    print(c)
  c = c+1
```

```python
# Look for all occurances of 6 and puts in a new_list
list_4 = [4,5,6,34,546,90,0.5,6,'monk']

c = 0
new_list = []
for i in list_4:
  if i == 6:
    print(c)
    new_list.append(c)
  c = c+1
new_list
```

## **3. Set - { }**

```python
# Set - does not accept duplicates
set_1 = {1,2,3,4,5,5}
set_1
```

```python
set_1 = {1,2,3,4,5,5}
set_2 = {5,6,2,8,9,9}
set_1.union(set_2)
```

```python
# what items are on both sets
set_1.intersection(set_2)
```

```python
# what differenciates set_1 that is not in set_2
set_1.difference(set_2)
```

```python
# what differenciates set_2 that is not in set_1
set_2.difference(set_1)
```

### **Defining & calling Function**

```python
# Define a function called greeting
def greeting():
  print("Happy New Year - 2026")  # the body does a print when called

# _ means: “I don’t care about the loop var”, repeat the foll code 3x, ignoring the counter
# can also use this where i is the loop var : for i in range(3):
# It’s a Python convention for unused variables
for _ in range(3):
  greeting()
```

```python
for i in range(2):    # Call the function greeting two times
  greeting()
```

```python
# same as above, Call the function 'greeting' two times
for _ in range(2):
  greeting()
```
