##5. Segregating the Dataset into Input and Output

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
x
```

##6. Splitting the dataset into Train and Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x.shape, y.shape
```

```python
x_train.shape,y_train.shape, x_test.shape,y_test.shape
```

##7. Loading the Dataset

```python
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier()
```

```python
model
```

##8. Training the Dataset

```python
model.fit(x_train,y_train)
```

##9. Predicting the data

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

##10. Calculate the accuracy of the trained model

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
