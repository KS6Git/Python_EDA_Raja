## **Heatmaps to identify complex relationships**


```python
# Heatmap

sns.heatmap(df_train.corr(numeric_only=True),annot=True)
```

```python
# Other Heatmaps from internet to analyse Survival rate to Embarked and Pclass
pivot = (
    df_train
    .dropna(subset=['Embarked'])
    .pivot_table(
        values='Survived',
        index='Pclass',
        columns='Embarked',
        aggfunc='mean'
    )
)

sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title('Survival Rate Heatmap')
plt.show()
```

##Ans: Heatmap shows there was more survival from Pclass-1 & Embark_C, and Pclass-2 & Embark-Q but generally more survivors in Pclass 1 & 2
